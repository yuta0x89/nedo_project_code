```python
import json
```

```python
with open("./cocotext.v2.json", "r", encoding="utf-8") as f:
    dataset = json.load(f)
```

```python
cnt = 0

str_list = []

for key, val in dataset["anns"].items():
    utf8_string = val["utf8_string"]
    image_id = val["image_id"]
    if utf8_string.strip().replace(" ", "") != "":
        cnt += 1
        str_list.append({"text": utf8_string, "file_name": dataset["imgs"][str(image_id)]["file_name"]})

print(cnt)
print(str_list[:10])
```
